### Correlation

```python
cols = [
    'case_ern_incm_amt', 'case_uern_incm_amt', 'case_slf_ern_incm_amt',
    'tot_shl_cost_amt', 'av_age', 'med_age', 'family_size',
    'max_educ', 'tt_income', 'unit_children',
    'homeless_allow_amt',
    'bas_utl_allow_amt', 'med_exp_amt', 'chld_sup_exp_amt',
    'pctchg_unearned_inc', 'pctchg_earned_inc', 'pctchg_shelter_cost',
    'aboveavg_uern_fips', 'aboveavg_ern_fips', 'aboveavg_shl_fips',
    'caseload', 'job_tenure', 'dss_tenure',
    'group_region', 'Issuance',
    'total_income', 'share_rent',
    'inc_cap', 'max_snap', 'share_issuance', 'new_case_num',
    'new_emp_num','Error', 'error_target'
    ]
```

```python
corr = df[cols].corr()

plt.figure(figsize=(25,23))
sns.heatmap(corr, annot=True)
plt.tight_layout()
plt.show()
```

```python
correlations = corr.unstack().sort_values(ascending = False)
correlations = correlations[correlations < 1.0].drop_duplicates()
correlations.head(10)
```

```python
import matplotlib.pyplot as plt

pairs = [
    ("family_size", "max_snap"),
    ("med_age", "av_age"),
    ("tt_income", "total_income"),
    ("unit_children", "family_size"),
    ("max_snap", "unit_children"),
    ("aboveavg_ern_fips", "case_ern_incm_amt"),
    ("case_uern_incm_amt", "aboveavg_uern_fips"),
    ("aboveavg_shl_fips", "tot_shl_cost_amt"),
    ("tt_income", "case_ern_incm_amt"),
    ("case_ern_incm_amt", "total_income"),
]

fig, axes = plt.subplots(2, 5, figsize=(22, 9))

for ax, (x, y) in zip(axes.flat, pairs):
    sc = ax.scatter(
        df[x], df[y],
        c=df["error_target"],
        cmap="coolwarm",
        alpha=0.5,
        s=15,
        edgecolors="none",
    )
    ax.set_xlabel(x)
    ax.set_ylabel(y)
    ax.set_title(f"{y} vs {x}")


handles, _ = sc.legend_elements()
fig.legend(handles, ["No error (0)", "Error (1)"],
           loc="upper right", title="error_target")

fig.suptitle("Top correlated pairs, colored by error_target", fontsize=14)
fig.tight_layout(rect=[0, 0, 1, 0.96])
plt.show();
```
